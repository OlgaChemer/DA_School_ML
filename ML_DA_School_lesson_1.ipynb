### Предсказать параметр MEDV из датасета housing.csv:
1 - модель линейной регрессии.

2 - разбить train/test на 0.8/0.2

```python
import numpy as np
import pandas as pd
from numpy.linalg import inv
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

```python
# Загружаем датасет
df = pd.read_csv('housing.csv', sep=',')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Задаем значения
X = df.iloc[:, :-1].copy()
y = df.iloc[:, -1].copy()
```

```python
# Разбиваем датасет на валидационные и тренировочные данные
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=10)

# Обучаем модель линейной регрессии
model = LinearRegression().fit(X_train, y_train)

# Предсказываем
y_pred = model.predict(X_valid)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# Проверяем качество метрики
print(r2_score(y_valid, y_pred))
```

```python
from sklearn.linear_model import Ridge, Lasso

# Проверим качество моделей Ridge, Lasso
model_Ridge = Ridge(alpha=0.8).fit(X_train, y_train)

# Предсказываем
y_pred_Ridge = model_Ridge.predict(X_valid)
y_pred_train_Ridge = model.predict(X_train)

print("Качество на валидации: %.3f" % mean_squared_error(y_valid, y_pred_Ridge))
print("Качество на обучении: %.3f" % mean_squared_error(y_train, y_pred_train_Ridge))
```

```python
model_Lasso = Lasso(alpha=0.8).fit(X_train, y_train)
y_pred_train_Ridge = model.predict(X_train)
y_pred_Lasso = model_Ridge.predict(X_valid)
y_pred_train_Lasso = model.predict(X_train)

print("Качество на валидации: %.3f" % mean_squared_error(y_valid, y_pred_Lasso))
print("Качество на обучении: %.3f" % mean_squared_error(y_train, y_pred_train_Lasso))
```
